```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset("iris")
print(df.shape)
df.head()
```

```python
print(df.describe())

print(df.info())
```

```python
df[df['sepal_length']==df['sepal_length'].max()]
```

```python
df.isna().sum()
```

```python
# Plot Graph

sns.pairplot(df, hue='species')

sns.pairplot(df, hue='species', diag_kind='hist')

sns.pairplot(df, hue='species', diag_kind='hist', palette='husl')

sns.pairplot(df, hue='species', diag_kind='hist', palette='husl', markers=['o','s','D'])
```

```python
# Line Graph

sns.lineplot(x='sepal_length', y='sepal_width', data=df)

sns.lineplot(x='sepal_length', y='sepal_width', data=df, hue='species')

sns.lineplot(x='sepal_length', y='sepal_width', data=df, hue='species', style='species')
```

```python
import matplotlib.pyplot as plt
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(df['sepal_length'])):
    ax.scatter(df['sepal_length'][i], df['sepal_width'][i])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
plt.show()
```
