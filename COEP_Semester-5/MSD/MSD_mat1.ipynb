```python
import matplotlib.pyplot as plt
```

```python
#Initalise the data
X=[10,20,30,40]
Y=[20,30,40,50]
```

```python
#plotting the data
plt.plot(X,Y)
#plt.show()
```

```python
#addithin the title
plt.title("Simple plot")
plt.plot(X,Y)
```

```python
#adding the labels
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.plot(X,Y)
```

```python
#add axis
plt.axis([0,60,0,60])
plt.plot(X,Y)
```

```python
Year=[2012,2014,2017,2019,2022]
India=[58.5, 70, 89.12, 102.13, 200]
Bangladesh=[49.3,65.5, 79.45, 102,167.70]
```

```python
plt.plot(Year,India,color='orange')
plt.show()
```

```python
plt.plot(Year,India,color='orange')
plt.plot(Year,Bangladesh,color='green')
plt.show()
```

```python
plt.plot(Year,India,color='orange',label="India")
plt.plot(Year,Bangladesh,color='green',label='Bangladesh')
plt.legend()
plt.show()
```

```python
plt.plot(Year,India,color='orange',marker='o',label="India")
plt.plot(Year,Bangladesh,color='green',label='Bangladesh')
plt.legend()
plt.show()
```

```python
from matplotlib.markers import MarkerStyle
plt.plot(Year,India,color='orange',marker='o',markersize=12,mec='b',label="India")
plt.plot(Year,Bangladesh,color='green',marker='d',mfc='r',label='Bangladesh')
plt.legend()
plt.show()
```

```python
from matplotlib.markers import MarkerStyle
plt.plot(Year,India,color='orange',marker='o',markersize=12,mec='b',mfc='r',label="India")
plt.plot(Year,Bangladesh,color='green',marker='d',linestyle='--',mfc='r',label='Bangladesh')
plt.legend()
plt.show()
```
